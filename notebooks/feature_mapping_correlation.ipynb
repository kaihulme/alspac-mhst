## Correlation

Now the categorical features have been mapped we can check the correlation between __all__ features.

```python
feature_corr = data_df.corr()
feature_corr
```

We can now visualise the correlations with a heatmap.

```python
figsave_dir = os.path.join(base_dir, "figures/feature_heatmap.png")

fig = plt.figure(figsize=(20, 20))
ax = sns.heatmap(feature_corr, cmap="coolwarm", vmin=-1, vmax=1)
plt.savefig(figsave_dir, bbox_inches="tight", facecolor="white")
```

Pandas `corr` is able to handle most of the missing values, but we can see a few of the features have no correlation between them.

This is likely due to there being no rows in the dataframe where the feature pairs both have a value.

We shall check this theory.

```python
nan_corr_features = ["comp_house", "tv_bed_16", "comp_bed_9", "mat_dep"]
# ncf_df = data_df[nan_corr_features].copy()
ncf_df = data_df.copy()

feature_pairs_with_na = []
for feature_x in ncf_df:
    for feature_y in ncf_df:
        if feature_x == feature_y:
            break
        pair_df = ncf_df[[feature_x, feature_y]].copy()
        has_non_nan = False
        print(f"[{feature_x}, {feature_y}]:")
        for i, row in pair_df.iterrows():
            if row[0] > 0 and row[1] > 0:
                print(f"non-NaN at row {i} ({row[0]}, {row[1]})\n")
                has_non_nan = True
                break
        if not has_non_nan:
            print("has no non-NaN row!\n")
            feature_pairs_with_na.append((feature_x, feature_y))

if feature_pairs_with_na:
    print(f"Feature pairs with no non-NaN rows:\n{feature_pairs_with_na}")
```

It seems for every pair there is at least one row where both values exist, so this is not the reason.

Another possibilty is that there is only one value that is non-NaN for that feature, lets check this:

```python
for col in data_df:
    unique = data_df[col].unique()
    unique_len = len(unique)
    if np.isnan(unique).any() and unique_len < 3:
        print(f"{col}: {unique_len} {unique}")
```

As we can see from this the only features that have only 1 unique value (exluding NaNs) are the ones which have no value in the correlation heatmap.
